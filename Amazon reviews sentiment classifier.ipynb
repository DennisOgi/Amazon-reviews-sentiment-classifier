### Importing libraries

```python
import numpy as np
import random
from sklearn.model_selection import train_test_split
import json
from sklearn.feature_extraction.text import CountVectorizer
```

### Creating data classes

```python
class Sentiment:
    Negative = 'NEGATIVE'
    Neutral = 'NEUTRAL'
    Positive = 'POSITIVE'
    
```

```python
class Review:
    def __init__(self, text, score):
        self.text = text
        self.score = score
        self.sentiment = self.get_sentiment()
        
    def get_sentiment(self):
        if self.score <= 2:
            return Sentiment.Negative
        elif self.score ==3:
            return Sentiment.Neutral
        else:
            return Sentiment.Positive
```

### Loading data

```python
file_name = 'Books_small_10000.json' #Changed the previous books_small.json data set to Books_small_10000.json data set to fix the data imbalance

reviews = []
with open(file_name) as f:
    for line in f:
        review = json.loads(line)
        reviews.append(Review(review['reviewText'], review['overall']))
        
        
reviews[5].text
```

### Splitting data into train and test

```python
train, test = train_test_split(reviews, test_size=0.3, random_state=42)
```

```python
print(train[0].sentiment)
```

### Seperating text data and sentiment score data

```python
x_train = [x.text for x in train]
y_train = [x.sentiment for x in train]

y_train
```

```python
x_train[0]
```

```python
y_train[0]
```

```python
x_test = [x.text for x in test]
y_test = [x.sentiment for x in test]

y_test
```

### Bag of words concept: using count vectorizer to transform text data into a format that the model can understand

```python
#Using count vectorizer to transform text data
vectorizer = CountVectorizer()

x_train_vec = vectorizer.fit_transform(x_train)

x_test_vectors = vectorizer.transform(x_test)
```

### Building our models

### Support Vector Machine

```python
#Svm classifier
from sklearn import svm
```

```python
clf_svm = svm.SVC(kernel='linear')
```

```python
clf_svm.fit(x_train_vec, y_train)
```

```python
x_test[0]
```

```python
clf_svm.predict(x_test_vectors)
```

### Decision Tree

```python
#Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_clf = DecisionTreeClassifier()

dt_clf.fit(x_train_vec, y_train)
```

```python
dt_clf.predict(x_test_vectors[0])
```

### Logistic Regression

```python
#Logistic regression classifier
from sklearn.linear_model import LogisticRegression
```

```python
log_clf = LogisticRegression(max_iter=1000)

log_clf.fit(x_train_vec, y_train)
```

```python
log_clf.predict(x_test_vectors[0])
```

### Random Forest

```python
#Random forest classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier()

rf_clf.fit(x_train_vec, y_train)
```

```python
rf_clf.predict(x_test_vectors[0])
```

### Evaluation

### Mean accuracy

```python
print(" The mean accuracy score of the svm classifier: {}".format(clf_svm.score(x_test_vectors, y_test)))
print(" The mean accuracy score of the decision tree classifier: {}".format(dt_clf.score(x_test_vectors, y_test)))
print(" The mean accuracy score of the logistic regression classifier: {}".format(log_clf.score(x_test_vectors, y_test)))
print(" The mean accuracy score of the random forest classifier: {}".format(rf_clf.score(x_test_vectors, y_test)))











```

### F1 score

```python
from sklearn.metrics import f1_score
```

```python
#Svm classifier
f1_score(y_test, clf_svm.predict(x_test_vectors), average=None, labels=[Sentiment.Positive,Sentiment.Negative])
```

```python
#Decision tree classifier
f1_score(y_test, dt_clf.predict(x_test_vectors), average=None, labels=[Sentiment.Positive,Sentiment.Negative])
```

```python
#Logistic regression classifier
f1_score(y_test, log_clf.predict(x_test_vectors), average=None, labels=[Sentiment.Positive,Sentiment.Negative])
```

```python
#Random forest classifier
f1_score(y_test, rf_clf.predict(x_test_vectors), average=None, labels=[Sentiment.Positive,Sentiment.Negative])
```
